```python
import random


# defining vector X will which will store which tag is chosen for sending messages.
# X[0] is by default 0 , for making it 1-based indexing.
# assuming 60% nodes are selected which want to transfer

def select_nodes():

    # Generate a list of 20 random values between 0 and 1
    X = [random.uniform(0, 1) for _ in range(num_points + 1)]

    # Replace values less than 0.4 with 0 and values greater than or equal to 0.4 with 1
    X = [0 if v < 0.4 else 1 for v in X]
    X[0] = 0
    #print(X)
    
    return X

```

```python

# function for selecting random destination for all transmitting nodes.

import numpy as np

def select_destinations(X_vec):
    # dest vector to store destination of selected transmitting nodes. 
    dest = [0 for i in range(num_points + 1)]

    # Setting up destination of all transmitting nodes.
    selected_nodes = [i for i, x in enumerate(X_vec) if x == 1]


    for node in selected_nodes:
        neighbors = np.array(neighbor_table.loc[f"P{node}"])
        neighbor_indices = np.where(neighbors == 1)[0]
        if len(neighbor_indices) > 0:
            random_neighbor = np.random.choice(neighbor_indices) + 1
            dest[node] = random_neighbor
        else:
            pass

    #print(dest)
    return dest

```

```python
# Function for defining C3 constraints.
# Note: here we are passing X again for vector m because assumption is that each nodes have single msg only.

def define_c3(x, m, dst, neighbor):
    
    # vector to store all constraints
    constraint = []
    
    for i in range(len(x)):
        str = [] # for storing constraint of a single tag
        
        # for every active tags we will see interfering tags in neighbor table and append them to the constraints.
        if(x[i] == 1):
            #str.append(f'g(c{i}{m[i]}, ')
            ind = dst[i]
            str.append(ind)
            for j in range(num_points):
                if(neighbor[f'P{ind}'][f'P{j + 1}'] == 1 and j != i - 1):
                    #str.append(f'x{j + 1} ')
                    #str.append('+ ')
                    str.append(j + 1)
             
            #if(str[-1] == '+ ' ): # just to remove extra + at the end
                #str = str[: -1]
                
            #str.append(f')')
            constraint.append(str)
        else:
            constraint.append('')
            
    return constraint
```

```python
# Function for solving a particular instance of selected transmitting nodes. 

import sys
import itertools
def solve_instance(x, d, c):
    
    indexes = [i for i, num in enumerate(x) if num == 1]
    #print(indexes)
    combinations = []
    result = None
    unique_combinations = []
    comb_size = -1

    for size in range(1, len(indexes) + 1):
        combinations = itertools.combinations(indexes, size)
        
        
        for combination in combinations:
            sorted_combination = tuple(sorted(combination))
            if sorted_combination not in unique_combinations:
                unique_combinations.append(sorted_combination)
                
                
                valid_comb = True
                for elem in sorted_combination:
                    intersection = set(c[elem]).intersection(set(sorted_combination))
                    if(len(intersection) > 0): 
                        valid_comb = False
                        break
                
                if(valid_comb):
                    if(comb_size < len(sorted_combination)):
                        result = sorted_combination
                
                    
    return list(result)
```

```python
# Utility function for converting solutions to binary array.

def list_to_binary_array(dframe):
    

    # length of binary array, i.e. equal to number of nodes.
    n = num_points

    # Create a new DataFrame with one binary feature for each possible index
    binary_cols = [f'label_{i+1}' for i in range(n)]
    binary_df = pd.DataFrame(np.zeros((len(dframe), n)), columns=binary_cols)

    # Set the appropriate binary features to 1 for each row and each list index
    for i, lst in enumerate(dframe['Solution']):
        for j in lst:
            binary_df.loc[i, f'label_{j}'] = 1

    # Concatenate the binary DataFrame with the original DataFrame
    result_df = pd.concat([dframe, binary_df], axis=1)

    # Drop the original list column if desired
    result_df.drop('Solution', axis=1, inplace=True)

    # Print the resulting DataFrame
    return result_df
```

```python
# function to do the preprocessing
import ast

def preprocess_data(tmpdata):
    tmpdata.dropna(inplace=True)  #dropping empty rows , if any
    tmpdata['Instance'] = tmpdata['Instance'].apply(ast.literal_eval)
    tmpdata['Solution'] = tmpdata['Solution'].apply(ast.literal_eval)
    data_X = tmpdata.drop(["Solution"], axis = 1)
    data_y = tmpdata.drop(["Instance"], axis = 1)
    split_df = data_X
    split_df = pd.DataFrame(split_df['Instance'].to_list(), columns=['Feature_1', 'Feature_2', 'Feature_3', 'Feature_4', 'Feature_5', 'Feature_6', 'Feature_7', 'Feature_8', 'Feature_9', 'Feature_10', 'Feature_11', 'Feature_12', 'Feature_13', 'Feature_14', 'Feature_15', 'Feature_16', 'Feature_17', 'Feature_18', 'Feature_19','Feature_20'])
    data_X = split_df
    #data_X = data_X.astype(int)

    # converiting solution list to binary target labels
    data_y = list_to_binary_array(data_y) 
    
    return data_X, data_y
```

```python
import random

# Define grid size and number of points
grid_size = 15
total_points = 20
percent_to_change = 0.25  # 25% of points to be changed
num_points_to_change = int(total_points * percent_to_change)

# Generate random points within the grid
random_points = [(random.randint(-grid_size // 2, grid_size // 2),
                  random.randint(-grid_size // 2, grid_size // 2))
                 for _ in range(total_points)]

# Randomly select points to change
points_to_change = random.sample(random_points, num_points_to_change)

# Function to generate a new position 2 units away in a random direction
def new_position(point):
    x, y = point
    directions = [(2, 0), (-2, 0), (0, 2), (0, -2)]
    dx, dy = random.choice(directions)
    return x + dx, y + dy

# Update the selected points with new positions
for i, point in enumerate(random_points):
    if point in points_to_change:
        new_pos = new_position(point)
        random_points[i] = new_pos

# Print the modified points
print("Original Points:")
print(random_points)

# Optionally, you can plot the points using a library like Matplotlib
import matplotlib.pyplot as plt

x, y = zip(*random_points)
plt.scatter(x, y, label="Modified Points", color='red')
plt.legend()
plt.grid()
plt.show()
```
